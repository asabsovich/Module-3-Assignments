**4.3 Answer:**  
To improve the predictive performance of the models, three additional features were constructed from existing variables to better capture relationships between property characteristics and housing value: living_area_ratio, room_density, and age.

The first engineered feature, living_area_ratio, was defined as the ratio of finished living area to total lot size (calculatedfinishedsquarefeet / lotsizesquarefeet). While both square footage and lot size are informative individually, this ratio provides a more meaningful measure of how efficiently a property’s land is utilized. Properties with similar lot sizes can differ significantly in value depending on how much of that space is developed into usable living area, making this a more context-aware representation than either variable alone.

The second feature, room_density, was calculated as (bedroomcnt + bathroomcnt) divided by calculatedfinishedsquarefeet. This transformation captures how densely rooms are distributed within a home. Two properties with the same square footage may have very different layouts depending on the number of rooms, and this feature helps quantify that distinction. By normalizing room count by size, the model gains insight into layout efficiency that is not directly observable from the original features.

The third feature, age, was computed as (2025 − yearbuilt). This transformation converts the original year built variable into a more interpretable measure of property age. Age serves as a proxy for factors such as depreciation, maintenance needs, and modernization, all of which influence property value. This representation aligns more directly with how buyers and appraisers assess housing value in practice.

After incorporating these engineered features, model performance improved consistently across all three regression models. Mean absolute error decreased from approximately $249,600 to $244,100, representing an improvement of about $5,500. This indicates that the engineered feature set provided additional predictive signal beyond the original variables.

It is important to interpret this result carefully. While performance improved after adding these features, the improvement reflects the combined effect of the engineered feature set, rather than the isolated impact of any single feature.
